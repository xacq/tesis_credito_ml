```python
# 02_analisis_exploratorio.ipynb
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Cargar datos originales (es mejor graficar con los nombres reales antes de codificar)
df = pd.read_csv('../data/datos_credito_simulados.csv')

# Configuración visual
sns.set_style("whitegrid")

```

```python
# GRÁFICO 1: Distribución de la Variable Objetivo (Balance de clases)
plt.figure(figsize=(6, 4))
sns.countplot(x='riesgo_real', data=df)
plt.title('Distribución de Clases: Buen Pagador (0) vs Alto Riesgo (1)')
plt.xlabel('Nivel de Riesgo')
plt.ylabel('Cantidad de Socios')
plt.savefig('figura_balance_clases.png') # Guarda la imagen para tu tesis
plt.show()

```

```python
# GRÁFICO 2: Relación entre Mora y Score Interno
plt.figure(figsize=(10, 6))
sns.boxplot(x='score_interno', y='dias_mora_prom', data=df, order=['AAA', 'AA', 'A', 'Analista', 'Rechazado'])
plt.title('Días de Mora Promedio según Calificación Interna')
plt.xlabel('Calificación Institucional')
plt.ylabel('Días de Mora')
plt.savefig('figura_mora_vs_score.png')
plt.show()

```

```python
# GRÁFICO 3: Matriz de Correlación (Mapa de Calor)
# Seleccionamos solo numéricas para ver correlaciones
cols_corr = ['dias_mora_prom', 'monto_solicitado', 'ingreso_mensual', 'edad', 'riesgo_real']
plt.figure(figsize=(8, 6))
sns.heatmap(df[cols_corr].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlación de Variables Numéricas')
plt.savefig('figura_correlacion.png')
plt.show()

```
