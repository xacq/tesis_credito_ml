```python
# 02_analisis_exploratorio.ipynb
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/datos_credito_simulados.csv")
```

```python
# -------------------------------
# 1. Distribución del riesgo
# -------------------------------
sns.countplot(x='riesgo_real', data=df)
plt.title("Distribución de Riesgo Crediticio")
plt.show()
```

```python
# -------------------------------
# 2. Score vs Riesgo
# -------------------------------
sns.barplot(x='score_interno', y='riesgo_real', data=df, order=['AAA','AA','A','Rechazado'])
plt.title("Riesgo promedio por Score Interno")
plt.show()

```

```python
# -------------------------------
# 3. Segmento vs Riesgo
# -------------------------------
sns.barplot(x='segmento_credito', y='riesgo_real', data=df)
plt.title("Riesgo promedio por Segmento Crediticio")
plt.show()
```

```python
# -------------------------------
# 4. Garantía vs Riesgo
# -------------------------------
sns.barplot(x='garantia', y='riesgo_real', data=df)
plt.title("Riesgo promedio según Tipo de Garantía")
plt.show()
```

```python
# -------------------------------
# 5. Estado legal vs Riesgo
# -------------------------------
sns.barplot(x='estado_legal', y='riesgo_real', data=df)
plt.title("Impacto del Estado Legal en el Riesgo")
plt.show()
```

```python
# -------------------------------
# 6. Propiedad completa vs Riesgo
# -------------------------------
sns.barplot(x='propiedad_completa', y='riesgo_real', data=df)
plt.title("Riesgo según Propiedad Completa del Bien")
plt.show()
```

```python
# Seleccionamos solo numéricas para ver correlaciones
cols_corr = ['dias_mora_prom', 'monto_solicitado', 'ingreso_mensual', 'edad', 'riesgo_real']
plt.figure(figsize=(8, 6))
sns.heatmap(df[cols_corr].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlación de Variables Numéricas')
plt.savefig('figura_correlacion.png')
plt.show()
```

```python
# Relación entre Mora y Score Interno
plt.figure(figsize=(10, 6))
sns.boxplot(x='score_interno', y='dias_mora_prom', data=df, order=['AAA', 'AA', 'A', 'Rechazado'])
plt.title('Días de Mora Promedio según Calificación Interna')
plt.xlabel('Calificación Institucional')
plt.ylabel('Días de Mora')
plt.savefig('figura_mora_vs_score.png')
plt.show()

```
